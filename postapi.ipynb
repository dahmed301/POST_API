```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Pump_Tests\Initial_Campaign\Test_2021_21-07-28_1558_PT_850rpm.lvm', sep='\t', skiprows=21)
```

```python
df.head()
```

```python
selected_columns = list(range(0, 9)) + list(range(21, 26))
df = df.iloc[:, selected_columns]
df.head()
```

```python
col_names = ['time','P1_bar','P2_bar','P3_bar','P4_bar','P5_bar','F1_lpermin','F2','RPM','dpv1','dpv2','dpv3','dpv4','dpv5']
```

```python
new_cols = dict(zip(df.columns,col_names))
df = df.rename(columns = new_cols)
df.columns
df.head()
```

```python
df[['RPM']] = df[['RPM']].astype(int)
df['RPM'].dtype
df.head()
```

```python
df.shape
```

```python
def select_data(row_num):
    content = df.loc[row_num,['P1_bar','P2_bar','P3_bar','P4_bar','P5_bar','RPM','F1_lpermin']]
    content= content.to_dict()
    return content
```

```python
def show_dpv(row_num):
    dpv = df.loc[row_num,['dpv1','dpv2','dpv3','dpv4','dpv5']]
    dpv = (dpv/160) - 25
    return dpv
```

```python
import requests
row_num = int(input("Enter a row number (0-1200): "))
aws = 'http://ec2-35-86-109-50.us-west-2.compute.amazonaws.com:8000'
r = requests.post(aws, json= select_data(row_num))
print(r.status_code)
response = r.json()
#print(response)
result = pd.DataFrame(list(response.items()), columns =['Response','Values'])
print(result.head(15))

show_dpv(row_num)
```
